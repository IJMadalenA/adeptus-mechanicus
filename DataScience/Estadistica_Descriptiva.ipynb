# Medidas de Relacion entre Variables.

- ## Covarianza
    - Es la estadistica principal para medir la relaci贸n entre dos variables.
    
    Covarianza poblacional:
    
    ### $$ \sigma_{xy} = \frac{{\sum \limits_{i = 1} ^{N}}(x_i - \mu_x)*(y_i - \mu_y)}{N} $$
    
    Covarianza muestral:
    
    ### $$ s_{xy} = \frac{{\sum \limits_{i = 1} ^{n}}(x_i - \bar{x})*(y_i - \bar{y})}{n-1} $$

```python
sns.scatterplot(data = df, x="reading score", y="writing score")
```

```python
x = df['reading score'].mean()
y = df['writing score'].mean()

x = df['reading score']
y = df['writing score']

# Formula para la covarianza poblacional.
_xy = np.sum([(i - x) * (j - y)for i, j in zip(x,y)]) / len(df)
_xy
```

```python
#Funci贸n para la matriz de covarianza muestral.
cov_mat = df.cov()
cov_mat
```
